```python
import numpy as np
from itertools import permutations, product
arr = np.random.randint(1, 11, 4)
arr = arr.tolist()
arr = [2, 3, 5 ,12]
```

```python
all_arr = list(set(permutations(arr, 4)))
signs = ['*', '+', '/', '-']
```

```python
def operate(arr):
    result = [[[arr[0], f'{arr[0]}']]]
    for k in range(1, 4):
        new_result = []
        cur_result = result[-1]
        for i in range(len(cur_result)):
            for j in range(len(signs)):
                if signs[j] == "+":
                    t = [cur_result[i][0] + arr[k],
                        f"{cur_result[i][1]} + {arr[k]}"]
                elif signs[j] == "-":
                    t = [cur_result[i][0] - arr[k],
                        f"{cur_result[i][1]} - {arr[k]}"]
                elif signs[j] == "*":
                    t = [cur_result[i][0] * arr[k],
                        f"{cur_result[i][1]} * {arr[k]}"]
                else:
                    t = [cur_result[i][0] / arr[k],
                        f"{cur_result[i][1]} / {arr[k]}"]
                new_result.append(t)
        result.append(new_result)
    for i in range(len(result[-1])):
        if result[-1][i][0] == 24:
            print(arr, result[-1][i])
```

```python
for arr in all_arr:
    operate(arr)
```

```python
from itertools import product

# Define variables
a = 2
b = 5
c = 5
d = 12

# Define all possible operator combinations
signs = ['*', '+', '/', '-']
# Generates all combinations of 3 operators
all_signs = list(product(signs, repeat=3))

# Original expressions (with * as placeholders)
expressions = [
    f"{a} * {b} * {c} * {d}",
    f"({a} * {b}) * {c} * {d}",
    f"{a} * ({b} * {c}) * {d}",
    f"{a} * {b} * ({c} * {d})",
    f"({a} * {b}) * ({c} * {d})",
    f"({a} * {b} * {c}) * {d}",
    f"{a} * ({b} * {c} * {d})",
    f"(({a} * {b}) * {c}) * {d}",
    f"{a} * (({b} * {c}) * {d})",
    f"({a} * ({b} * {c})) * {d}",
    f"{a} * ({b} * ({c} * {d}))"
]

# Replace * with all_signs[0], all_signs[1], etc.
all_results = []
for i, ops in enumerate(all_signs):
    for expr in expressions:
        # Replace * with the current operator combination
        modified_expr = expr.replace('*', '{}').format(*ops)
        try:
            # Evaluate the modified expression
            result = eval(modified_expr)
            all_results.append(f"{modified_expr} = {result}")
        except ZeroDivisionError:
            print(f"{modified_expr} = Division by zero error")
```
